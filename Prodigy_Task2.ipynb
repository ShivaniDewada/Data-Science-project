```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
data = pd.read_csv('titanic.csv')
```

```python
# Display the first few rows of the dataset
print(data.head())

# Check the summary statistics of the dataset
print(data.describe())

# Check the data types and missing values
print(data.info())
```

```python
# Compute the correlation matrix
correlation_matrix = data.corr()

# Plot the correlation heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Plot the count of survivors vs. non-survivors
sns.countplot(x='Survived', data=data)
plt.title('Count of Survivors vs. Non-Survivors')
plt.show()

# Plot the count of survivors based on gender
sns.countplot(x='Survived', hue='Sex', data=data)
plt.title('Count of Survivors based on Gender')
plt.show()

# Plot the count of survivors based on passenger class
sns.countplot(x='Survived', hue='Pclass', data=data)
plt.title('Count of Survivors based on Passenger Class')
plt.show()

# Plot the distribution of age among survivors and non-survivors
sns.histplot(data=data, x='Age', hue='Survived', kde=True)
plt.title('Distribution of Age among Survivors and Non-Survivors')
plt.show()

# Plot the count of passengers based on the port of embarkation
sns.countplot(x='Embarked', data=data)
plt.title('Count of Passengers based on the Port of Embarkation')
plt.show()

# Plot the count of survivors based on the port of embarkation
sns.countplot(x='Survived', hue='Embarked', data=data)
plt.title('Count of Survivors based on the Port of Embarkation')
plt.show()
```
